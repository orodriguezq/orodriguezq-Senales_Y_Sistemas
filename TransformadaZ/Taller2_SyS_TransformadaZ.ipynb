## Semejanzas y Diferencias entre:
### Serie de Fourier, Transformada de Fourier (tiempo continuo y discreto), Transformada de Laplace y Transformada Z

---

### 1. Serie de Fourier

- **Descripción:**  
  Representa señales **periódicas** como la suma infinita de senos y cosenos (o exponenciales complejas) de diferentes frecuencias.
- **Dominio:**  
  Tiempo continuo, señales periódicas.
- **Forma general:**  
  $$
  x(t) = \sum_{n=-\infty}^{\infty} C_n e^{j n \omega_0 t}
  $$
  donde $C_n$ son los coeficientes de la serie y $\omega_0$ es la frecuencia fundamental.

---

### 2. Transformada de Fourier en Tiempo Continuo (FT)

- **Descripción:**  
  Permite analizar señales **no necesariamente periódicas** descomponiéndolas en sus componentes de frecuencia.
- **Dominio:**  
  Tiempo continuo, señales aperiódicas y periódicas.
- **Forma general:**  
  $$
  X(f) = \int_{-\infty}^{\infty} x(t) e^{-j 2\pi f t} dt
  $$
  donde $X(f)$ es la representación en frecuencia de $x(t)$.

---

### 3. Transformada de Fourier en Tiempo Discreto (DTFT y DFT)

- **DTFT (Transformada de Fourier de Tiempo Discreto):**
  - **Descripción:**  
    Análisis de señales **discretas** en el tiempo, obteniendo una representación de frecuencia continua y periódica.
  - **Forma general:**  
    $$
    X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j\omega n}
    $$
- **DFT (Transformada Discreta de Fourier):**
  - **Descripción:**  
    Versión computacional (finita) de la DTFT, utilizada en algoritmos como la FFT.
  - **Dominio:**  
    Tiempo discreto, señales finitas.
  - **Forma general:**  
    $$
    X[k] = \sum_{n=0}^{N-1} x[n] e^{-j \frac{2\pi}{N} k n}
    $$

---

### 4. Transformada de Laplace

- **Descripción:**  
  Generaliza la transformada de Fourier usando un parámetro complejo $s$, permitiendo el análisis de estabilidad y de sistemas lineales, incluso con señales no absolutamente integrables o que crecen exponencialmente.
- **Dominio:**  
  Tiempo continuo, señales causales (principalmente).
- **Forma general:**  
  $$
  X(s) = \int_{-\infty}^{\infty} x(t) e^{-s t} dt
  $$
  donde $s = \sigma + j\omega$.

---

### 5. Transformada Z

- **Descripción:**  
  Analiza señales **discretas** y sistemas, generalizando la DTFT y permitiendo el estudio de estabilidad y respuesta en sistemas digitales.
- **Dominio:**  
  Tiempo discreto.
- **Forma general:**  
  $$
  X(z) = \sum_{n=-\infty}^{\infty} x[n] z^{-n}
  $$
  donde $z$ es un número complejo.

---

## Semejanzas

- Todas permiten **analizar señales** y sistemas en el dominio de la frecuencia o el plano complejo.
- La **Serie de Fourier**, **Transformada de Fourier** y **DTFT** descomponen señales en sumas/integrales de senos, cosenos o exponenciales.
- **Transformada de Laplace** y **Transformada Z** generalizan a Fourier y DTFT respectivamente, permitiendo analizar estabilidad y respuesta transitoria en sistemas (usando polos y ceros).
- Se usan ampliamente en procesamiento de señales, control, comunicaciones y análisis de sistemas lineales.

---

## Diferencias

| Concepto                         | Serie de Fourier             | Transformada de Fourier        | Laplace                      | Z                           |
|-----------------------------------|------------------------------|-------------------------------|------------------------------|-----------------------------|
| Dominio de aplicación            | Tiempo continuo periódico    | Tiempo continuo aperiódico    | Tiempo continuo              | Tiempo discreto             |
| Tipo de señal                    | Periódica                    | Aperiódica/periódica          | Causal/no causal             | Causal/no causal            |
| Uso de polos y ceros             | No                           | No                            | Sí                           | Sí                          |
| Plano de análisis                | Eje jω (frecuencia)          | Eje jω                        | Plano s = σ + jω             | Plano z                     |
| Estudio de estabilidad           | No                           | No                            | Sí                           | Sí                          |
| Generaliza a                    | -                            | Serie de Fourier (caso periódica) | Fourier (para s = jω)        | DTFT (para z = e^{jω})      |
| Cálculo                          | Suma infinita                | Integral                      | Integral (exponencial)       | Suma (exponencial discreta) |

---

## Resumen gráfico de relaciones

- **Serie de Fourier** ⟶ **Transformada de Fourier** (cuando la señal es aperiódica)
- **Transformada de Fourier** ⟶ **Laplace** (FT es un caso particular de Laplace para $s = j\omega$)
- **DTFT** ⟶ **Transformada Z** (DTFT es un caso particular de Z para $z = e^{j\omega}$)

```python
from sympy import symbols, summation, oo, simplify, Function, Poly
import numpy as np
import matplotlib.pyplot as plt

# Definir variables simbólicas
z, a, N, n = symbols('z a N n', complex=True)

# =============== CASO i) ====================
# x[n] = -a^n * u[-n-1]
expr1 = -a**n * z**(-n)
sum1 = summation(expr1, (n, -oo, -1))
sum1_simpl = simplify(sum1)

print("CASO i)\n------")
print("Transformada Z simbólica:")
print("Suma:", sum1)
print("Forma cerrada:", sum1_simpl)
print()

# =============== CASO ii) ===================
# x[n] = a^n para 0 <= n <= N-1
expr2 = a**n * z**(-n)
sum2 = summation(expr2, (n, 0, N-1))
sum2_simpl = simplify(sum2)

print("CASO ii)\n-------")
print("Transformada Z simbólica:")
print("Suma:", sum2)
print("Forma cerrada:", sum2_simpl)
print()

# =============== CASO iii) ==================
# x[n] = {5, 3, -2, 0, 4, -3}
x3 = [5, 3, -2, 0, 4, -3]
X3_z = sum(x3[n]*z**(-n) for n in range(len(x3)))

print("CASO iii)\n--------")
print("Transformada Z simbólica:")
print("Polinomio:", X3_z)
print("Polinomio en z^-1:", Poly(X3_z, z**-1))
print()

# ============ GRAFICAR RESPUESTA EN FRECUENCIA =============
# Valores numéricos a usar:
a_num = 0.8
N_num = 6

# CASO i
def X1_z_num(zv):
    return -zv / (a_num - zv)

# CASO ii
def X2_z_num(zv):
    azinv = a_num / zv
    return (1 - azinv**N_num) / (1 - azinv)

# CASO iii
def X3_z_num(zv):
    return sum(x3[n]*zv**(-n) for n in range(len(x3)))

# Círculo unitario
w = np.linspace(0, 2*np.pi, 400)
z_circ = np.exp(1j*w)

X1_mod = np.abs([X1_z_num(zv) for zv in z_circ])
X2_mod = np.abs([X2_z_num(zv) for zv in z_circ])
X3_mod = np.abs([X3_z_num(zv) for zv in z_circ])

# Graficar
plt.figure(figsize=(12,8))

plt.subplot(3,1,1)
plt.plot(w, X1_mod)
plt.title(r'Caso i: $x[n] = -a^n u[-n-1]$  |  $a = %.1f$' % a_num)
plt.ylabel(r'$|X_1(e^{j\omega})|$')
plt.grid()

plt.subplot(3,1,2)
plt.plot(w, X2_mod)
plt.title(r'Caso ii: $x[n] = a^n,\, 0 \leq n < N$  |  $a = %.1f$, $N = %d$' % (a_num, N_num))
plt.ylabel(r'$|X_2(e^{j\omega})|$')
plt.grid()

plt.subplot(3,1,3)
plt.plot(w, X3_mod)
plt.title(r'Caso iii: $x[n] = \{5, 3, -2, 0, 4, -3\}$')
plt.xlabel(r'$\omega$ [rad]')
plt.ylabel(r'$|X_3(e^{j\omega})|$')
plt.grid()

plt.tight_layout()
plt.show()
```

## Filtros de Respuesta al Impulso Finita (FIR) y Respuesta al Impulso Infinita (IIR)

### ¿Qué es un filtro FIR?

- **FIR** significa **Finite Impulse Response** o **Respuesta al Impulso Finita**.
- Un filtro FIR es un sistema lineal e invariante en el tiempo cuya **respuesta al impulso** (la salida cuando la entrada es un impulso unitario) **dura un número finito de muestras** y luego se hace exactamente cero.
- Su ecuación de diferencias es:
  $$
  y[n] = b_0 x[n] + b_1 x[n-1] + \cdots + b_M x[n-M]
  $$
  donde solo aparecen términos de entrada, **no hay realimentación** de salidas anteriores.
- Son **siempre estables** y pueden diseñarse con **fase lineal** (importante para evitar distorsión de fase).

---

### ¿Qué es un filtro IIR?

- **IIR** significa **Infinite Impulse Response** o **Respuesta al Impulso Infinita**.
- Un filtro IIR es un sistema cuya respuesta al impulso **nunca se hace exactamente cero**: puede durar para siempre (o tender a cero en el infinito).
- Su ecuación de diferencias es:
  $$
  y[n] = b_0 x[n] + b_1 x[n-1] + \cdots + b_M x[n-M] - a_1 y[n-1] - a_2 y[n-2] - \cdots - a_N y[n-N]
  $$
  donde **sí hay realimentación**: la salida depende de salidas pasadas.
- Los IIR pueden ser **inestables** si no se diseñan adecuadamente, pero suelen requerir **menos coeficientes** que un FIR para obtener una pendiente similar.
- Permiten implementar filtros digitales similares a los analógicos clásicos (Butterworth, Chebyshev, etc).

---

### Sistemas con y sin realimentación

- **Con realimentación**:
  - Son los sistemas donde la salida se usa como parte de la entrada futura.
  - En ecuaciones, se reconoce porque $y[n]$ depende de $y[n-1], y[n-2], ...$ (como los IIR).
  - Pueden tener respuesta infinita y ser inestables si los polos están fuera del círculo unitario (en Z).
- **Sin realimentación**:
  - La salida depende **solo** de la entrada presente y pasada, **no** de la salida anterior.
  - Todos los FIR son sistemas sin realimentación.
  - Siempre estables y con respuesta finita.

---

### **Paralelo entre FIR, IIR y realimentación**

| Característica         | Filtros FIR               | Filtros IIR               | Sistemas sin realimentación | Sistemas con realimentación |
|-----------------------|---------------------------|---------------------------|-----------------------------|-----------------------------|
| Respuesta al impulso  | Finita                    | Infinita                  | Finita                      | Infinita (en general)       |
| Ecuación de diferencias | Solo entradas ($x[n-k]$) | Entradas y salidas pasadas | Solo entradas               | Incluye salidas anteriores  |
| Estabilidad           | Siempre                   | No siempre                | Siempre                     | Depende de los polos        |
| Fase lineal           | Puede lograrse            | Difícil de lograr         | Sí                          | No necesariamente           |
| Uso de realimentación | No                        | Sí                        | No                          | Sí                          |
| Complejidad           | Orden más alto necesario  | Orden más bajo suficiente | Orden más alto necesario    | Orden más bajo suficiente   |
| Diseño típico         | Ventaneo, Parks-McClellan | Transformadas analógicas  | FIR                         | IIR                         |

---

### **Resumen visual:**
- **FIR = sistema sin realimentación = respuesta finita.**
- **IIR = sistema con realimentación = respuesta infinita.**

## Comparación entre estrategias de diseño: Invarianza al impulso vs Transformación bilineal

### **1. Invarianza al Impulso (Impulse Invariance Method)**

- **¿En qué consiste?**
  - Es un método para convertir filtros analógicos a digitales.
  - La idea es que la respuesta al impulso del filtro digital sea una muestra directa de la respuesta al impulso del filtro analógico.
  - Se obtiene la respuesta al impulso \( h_a(t) \) del sistema analógico y luego se muestrea: \( h[n] = h_a(nT) \), donde \( T \) es el periodo de muestreo.
- **Ventajas:**
  - Reproduce con exactitud los polos del sistema analógico dentro del círculo unitario.
  - Es fácil de implementar para sistemas cuyas respuestas al impulso son conocidas o simples.
  - Mantiene la forma de la respuesta (en términos de polos) en la versión digital.
- **Desventajas:**
  - **Problema de aliasing**: la respuesta en frecuencia del filtro digital puede contener "copias" de la respuesta analógica (por la periodicidad inherente al muestreo).
  - No es adecuada para filtros pasa altos o de banda ancha (el aliasing puede ser severo).
  - El mapeo de la frecuencia no es exacto, especialmente para frecuencias altas.

---

### **2. Transformación Bilineal (Bilinear Transformation Method)**

- **¿En qué consiste?**
  - Es otro método para convertir un filtro analógico a digital, mediante una transformación matemática que mapea todo el eje \( j\Omega \) (frecuencia) del plano \( s \) (analógico) al eje \( e^{j\omega} \) (digital).
  - La transformación típica es:  
    $$
    s = \frac{2}{T} \frac{1 - z^{-1}}{1 + z^{-1}}
    $$
  - Este mapeo garantiza que todo el eje de estabilidad analógico (lado izquierdo del plano s) se mapea dentro del círculo unitario digital.
- **Ventajas:**
  - **No hay aliasing:** la transformación es unívoca (no se superponen frecuencias).
  - Es posible controlar con precisión la frecuencia de corte deseada (aunque hay que pre-warping).
  - Preserva la estabilidad: un sistema estable analógico produce un sistema estable digital.
- **Desventajas:**
  - La relación de frecuencias entre el filtro analógico y el digital es **no lineal** (efecto de "warping" de frecuencia).
  - Hay que aplicar una **corrección de pre-warping** al especificar las frecuencias de diseño.
  - Puede distorsionar la forma de la respuesta en frecuencia en bandas altas.

---

### **Cuadro comparativo**

| Característica                  | Invarianza al Impulso                   | Transformación Bilineal               |
|----------------------------------|-----------------------------------------|---------------------------------------|
| ¿Hay aliasing?                   | Sí, puede haber aliasing                | No, no hay aliasing                   |
| ¿Se necesita pre-warping?        | No                                      | Sí, para ajustar frecuencias de corte  |
| ¿Mapea todo eje jΩ (s) al eje ω? | No (aliasing por muestreo)              | Sí, mapeo uno a uno                   |
| Precisión en frecuencia          | Buena en bajas frecuencias              | No lineal, mejor control con pre-warping |
| Complejidad                      | Más simple para algunos casos           | Ligero aumento de cálculo matemático  |
| Uso recomendado                  | Filtros pasa bajos, polos reales        | Filtros en general, todo tipo         |
| Preserva estabilidad             | Sí                                      | Sí                                    |
| Polos/ceros                     | Solo los polos se mapean directamente   | Polos y ceros se mapean según la transformación |

---

### **Resumen visual**
- **Invarianza al impulso:** simple, puede sufrir aliasing, fácil de implementar para algunos casos, adecuada para frecuencias bajas.
- **Transformación bilineal:** evita aliasing, requiere pre-warping, útil para todo tipo de filtros, mejor control de las frecuencias de corte.

```python
%%bash
cat > app.py << 'EOF'




import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import io
import os
from scipy.io import wavfile
from yt_dlp import YoutubeDL
from pydub import AudioSegment
import traceback

# --- Configuración de la app ---
st.set_page_config(page_title="AM Modulación/Demodulación", page_icon="📻", layout="wide")
st.title("AM Modulación y Demodulación Coherente")

# --- Descargar audio de YouTube ---
@st.cache_data
def download_youtube_audio(url: str, out_fname: str = 'audio.wav'):
    for f in os.listdir():
        if f.startswith('audio.') or f == out_fname:
            try: os.remove(f)
            except: pass
    opts = {'format':'bestaudio/best','outtmpl':'audio.%(ext)s','quiet':True}
    with YoutubeDL(opts) as ydl:
        info = ydl.extract_info(url, download=True)
        ext  = info.get('ext','m4a')
        fname= f"audio.{ext}"
    audio = AudioSegment.from_file(fname).set_channels(1)
    audio.export(out_fname, format='wav')
    os.remove(fname)
    return out_fname

# --- Dominio tiempo & frecuencia ---
def plot_time_freq(x, fs, title, t_lim=None, f_lim=None):
    N = len(x); t = np.arange(N)/fs
    fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,3))
    ax1.plot(t,x); ax1.set(xlabel='t [s]',ylabel='x(t)',title=f'{title} – Tiempo')
    if t_lim: ax1.set_xlim(t_lim)
    Xf = np.fft.fft(x); freqs = np.fft.fftfreq(N,1/fs)
    ax2.plot(freqs,np.abs(Xf)/N); ax2.set(xlabel='f [Hz]',ylabel='|X(f)|',title=f'{title} – Frecuencia')
    if f_lim: ax2.set_xlim(f_lim)
    fig.tight_layout(); st.pyplot(fig)

# --- Parámetros ---
st.sidebar.header("Parámetros señal y AM")
url = st.sidebar.text_input("URL de YouTube","https://www.youtube.com/watch?v=dQw4w9WgXcQ")
t0  = st.sidebar.number_input("Tiempo inicio (s)",min_value=0.0,value=0.0,step=0.1)
t1  = st.sidebar.number_input("Tiempo fin    (s)",min_value=0.1,value=1.0,step=0.1)
f0  = st.sidebar.number_input("Portadora f0 (Hz)",min_value=100.0,value=10000.0,step=100.0)
B   = st.sidebar.number_input("LPF ancho B (Hz)",min_value=100.0,value=5000.0,step=100.0)

# --- Parámetros filtro LPF ---
st.sidebar.header("Filtro pasabajas (LPF) Demodulación")
tipo_filtro = st.sidebar.selectbox("Tipo de filtro", ["FIR", "IIR"])
orden = st.sidebar.slider("Orden del filtro", min_value=2, max_value=10, value=5)

if tipo_filtro == "IIR":
    arch = st.sidebar.selectbox("Arquitectura IIR", ["Butterworth", "Chebyshev1", "Bessel"])
    if arch == "Chebyshev1":
        ripple = st.sidebar.slider("Ripple (dB, Chebyshev1)", 0.01, 2.0, 1.0)
else:
    arch = "FIR"
    ventana = st.sidebar.selectbox("Ventana FIR", ["hamming", "hann", "blackman", "boxcar"])

if url:
    try:
        # descarga y WAV mono
        wav_file = download_youtube_audio(url)
        fs,data = wavfile.read(wav_file)
        if data.ndim>1: data=data[:,0]
        audio_seg = AudioSegment.from_wav(wav_file).set_channels(1)

        # 1) mensaje
        ms0,ms1 = int(t0*1000),int(t1*1000)
        msg = audio_seg[ms0:ms1]
        m   = np.array(msg.get_array_of_samples()).astype(np.float32)
        Ac=np.max(np.abs(m))
        st.subheader("1) Mensaje m(t)")
        plot_time_freq(m,fs,'Mensaje',t_lim=(t0,t1),f_lim=(-B,B))
        buf=io.BytesIO(); msg.export(buf,format='wav'); buf.seek(0)
        st.audio(buf.read(),format='audio/wav',sample_rate=fs)

        # 2) portadora
        t=np.arange(len(m))/fs; carrier=np.cos(2*np.pi*f0*t)
        st.subheader("2) Portadora cos(2πf0t)")
        plot_time_freq(carrier,fs,'Portadora',t_lim=(t0,t1),f_lim=(-2*f0,2*f0))

        # 3) AM
        r=(1+m/Ac)*carrier
        st.subheader("3) Señal AM r(t)")
        plot_time_freq(r,fs,'AM',t_lim=(t0,t1),f_lim=(-2*f0,2*f0))
        r_int=(r/np.max(np.abs(r))*32767).astype(np.int16)
        r_seg=AudioSegment(r_int.tobytes(),sample_width=2,frame_rate=fs,channels=1)
        buf=io.BytesIO(); r_seg.export(buf,format='wav'); buf.seek(0)
        st.audio(buf.read(),format='audio/wav',sample_rate=fs)

        # 4) mezclador y1
        y1=r*carrier
        st.subheader("4) Mezclador y1(t)")
        plot_time_freq(y1,fs,'Mezclador',t_lim=(t0,t1),f_lim=(-2*f0,2*f0))

        # 5) FILTRADO LPF (personalizado)
        from scipy.signal import butter, cheby1, bessel, firwin, lfilter, freqz, tf2zpk

        nyq = fs/2
        fc = B / nyq

        if tipo_filtro == "FIR":
            taps = firwin(numtaps=orden+1, cutoff=fc, window=ventana)
            y2 = lfilter(taps, 1.0, y1)
            w, h = freqz(taps, worN=8000)
            b, a = taps, [1]
        else:
            if arch == "Butterworth":
                b, a = butter(orden, fc)
            elif arch == "Chebyshev1":
                b, a = cheby1(orden, ripple, fc)
            elif arch == "Bessel":
                b, a = bessel(orden, fc, norm='phase')
            y2 = lfilter(b, a, y1)
            w, h = freqz(b, a, worN=8000)

        st.subheader("5) Tras LPF y2(t)")
        plot_time_freq(y2,fs,'LPF',t_lim=(t0,t1),f_lim=(-B,B))
        y2_int=(y2/np.max(np.abs(y2))*32767).astype(np.int16)
        y2_seg=AudioSegment(y2_int.tobytes(),sample_width=2,frame_rate=fs,channels=1)
        buf=io.BytesIO(); y2_seg.export(buf,format='wav'); buf.seek(0)
        st.audio(buf.read(),format='audio/wav',sample_rate=fs)

        # 6) mensaje recuperado
        m_rec=(2/Ac)*y2
        st.subheader("6) Mensaje recuperado m_rec(t)")
        plot_time_freq(m_rec,fs,'Recuperado',t_lim=(t0,t1),f_lim=(-B,B))
        m_int=(m_rec/np.max(np.abs(m_rec))*32767).astype(np.int16)
        m_seg=AudioSegment(m_int.tobytes(),sample_width=2,frame_rate=fs,channels=1)
        buf=io.BytesIO(); m_seg.export(buf,format='wav'); buf.seek(0)
        st.audio(buf.read(),format='audio/wav',sample_rate=fs)

        # 7) DIAGRAMA DE BODE
        st.subheader("🟦 Diagrama de Bode del filtro implementado")
        fig, ax = plt.subplots(2, 1, figsize=(8, 5))
        ax[0].plot(0.5*fs*w/np.pi, 20*np.log10(np.abs(h)+1e-10))
        ax[0].set_title('Magnitud [dB]')
        ax[0].set_ylabel("Magnitud [dB]")
        ax[0].set_xlabel("Frecuencia [Hz]")
        ax[0].grid()
        ax[1].plot(0.5*fs*w/np.pi, np.unwrap(np.angle(h))*180/np.pi)
        ax[1].set_title('Fase [deg]')
        ax[1].set_ylabel("Fase [°]")
        ax[1].set_xlabel("Frecuencia [Hz]")
        ax[1].grid()
        fig.tight_layout()
        st.pyplot(fig)

        # 8) DIAGRAMA DE POLOS Y CEROS
        st.subheader("🟦 Diagrama de polos y ceros del filtro")
        z, p, k = tf2zpk(b, a)
        fig2, ax2 = plt.subplots(figsize=(4, 4))
        circle = plt.Circle((0, 0), 1, color='black', fill=False, linestyle='--')
        ax2.add_artist(circle)
        ax2.scatter(np.real(z), np.imag(z), marker='o', facecolors='none', edgecolors='b', label='Ceros')
        ax2.scatter(np.real(p), np.imag(p), marker='x', color='r', label='Polos')
        ax2.set_title("Polos y ceros")
        ax2.set_xlabel("Re")
        ax2.set_ylabel("Im")
        ax2.legend()
        ax2.grid()
        ax2.axis('equal')
        ax2.set_xlim(-1.5, 1.5)
        ax2.set_ylim(-1.5, 1.5)
        st.pyplot(fig2)

    except Exception as ex:
        st.error(f"Error en pipeline: {ex}")
        st.text(traceback.format_exc())
else:
    st.info("👉 Ingresa URL o ID en la barra lateral.")
EOF
```
